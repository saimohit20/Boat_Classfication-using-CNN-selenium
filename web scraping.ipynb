```python
chromePath = "D:\\Tools and programming-DS\\scrap images\\chromedriver.exe"
service = Service(chromePath)
driver = webdriver.Chrome(service=service)
search_URL = "https://www.google.com/search?sca_esv=41165fec71bc817d&sxsrf=ADLYWIJUAQXJTZjMF62j-4MOt0dJ88iNIQ:1732716237606&q=cape+may+ferry+sinking&uds=ADvngMjcH0KdF7qGWtwTBrP0nt7d0cdZ5RSnXmyZGDVCbHG77B9Q0DjkYts82JFQlgqXE5vIYqwDbjzlqLiESC_m5UYMjMwdmcdLK-B4Z0OsCTTFKEeDbrfkz0jNfW6Qi1md0GeS0RVz2PdqgZgkGFDC7RejUTiXoA&udm=2&sa=X&ved=2ahUKEwjh5Nr-1vyJAxW2AdsEHfR9BHEQxKsJegUIlAEQAQ&ictx=0&biw=1707&bih=811&dpr=1.13"
driver.get(search_URL)
try:
    reject_button = WebDriverWait(driver, 10).until(
        EC.element_to_be_clickable((By.XPATH, '//*[@id="W0wltc"]'))
    )
    reject_button.click()
    print("Clicked 'Alle ablehnen'.")
except Exception as e:
    print("Failed to find or click the 'Alle ablehnen' button:", e)
print(driver.title)
a = input("Waiting...")

folder_name = 'sink ships images'
if not os.path.isdir(folder_name):
    os.makedirs(folder_name)

def download_image(url, folder_name, num):

    reponse = requests.get(url)
    if reponse.status_code==200:
        with open(os.path.join(folder_name, str(num)+".jpg"), 'wb') as file:
            file.write(reponse.content)

driver.execute_script("window.scrollTo(0, 0);")
page_html = driver.page_source
pageSoup = bs4.BeautifulSoup(page_html, 'html.parser')
containers = pageSoup.findAll('div', {'class':"eA0Zlc WghbWd FnEtTd mkpRId m3LIae RLdvSe qyKxnc ivg-i PZPZlf GMCzAd"} )
print(len(containers))
len_containers = len(containers)
imageurl = []
for i in range(1, len_containers+1):
    if ((i % 25 == 0) | (i==18)):
        continue

    xPath = """//*[@id="rso"]/div/div/div[1]/div/div/div[%s]"""%(i)
    driver.find_element(By.XPATH, xPath).click()
    timeStarted = time.time()
    while True:
        imageElement = driver.find_element(By.XPATH,"""//*[@id="Sva75c"]/div[2]/div[2]/div/div[2]/c-wiz/div/div[3]/div[1]/a/img[1]""")
        imageURL= imageElement.get_attribute('src')
        currentTime = time.time()
        imageurl.append(imageURL)
        if currentTime - timeStarted > 10:
                print("Timeout! Will download a lower resolution image and move onto the next one")
                break
        try:
            download_image(imageURL, folder_name, f"sink_cap{i}")
            print("Downloaded element %s out of %s total. URL: %s" % (i, len_containers + 1, imageURL))
            break
        except:
            print("Couldn't download an image %s, continuing downloading the next one"%(i))
driver.quit()
```
